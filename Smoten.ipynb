```python
import pandas as pd
from imblearn.over_sampling import SMOTEN
from sklearn.preprocessing import LabelEncoder

# Load the dataset
data = pd.read_csv('adult.csv')
selected_columns = ["age", "native_country", "race", "sex", "income"]
data = data[selected_columns]

# Define the z-score threshold
z_score_threshold = 3

# Calculate z-scores for the 'age' column
z_scores = (data['age'] - data['age'].mean()) / data['age'].std()

# Identify outliers based on z-scores
outliers_age = data[abs(z_scores) > z_score_threshold]

# Remove outliers from the DataFrame
data_without_outliers_in_age = data.drop(outliers_age.index)
```

```python
# label_encoder = LabelEncoder()
# for column in ['income', 'native_country', 'sex', 'race']:
#     data_without_outliers_in_age[column] = label_encoder.fit_transform(data_without_outliers_in_age[column])


# Separate features and target variable for SMOTE
X_race = data_without_outliers_in_age.drop(columns=['race'])
y_race = data_without_outliers_in_age['race']

```

```python
# Apply SMOTE to the minority classes in 'race' column
sm_race = SMOTEN(random_state=42)
X_res_race, y_res_race = sm_race.fit_resample(X_race, y_race)

# Create a DataFrame with augmented data
augmented_data_race = pd.DataFrame(X_res_race, columns=X_race.columns)
print(augmented_data_race)

augmented_data_race['race'] = y_res_race
combined_data = pd.concat([data_without_outliers_in_age, augmented_data_race], ignore_index=True)
combined_data.to_csv('augmented_smoten.csv', index=False)

```
